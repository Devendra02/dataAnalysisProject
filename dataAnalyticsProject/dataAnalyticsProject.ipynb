# Exploratory Data Analysis of Shark Tank India Data 
I will be using libraries of Python like Numpy, Pandas, Matplotlib and Seaborn for analyzing the data.
**Numpy** is used for mathematical and statistical computation.

All combined together help in good analysis of Data. 

##### Shark Tank India is a popular reality TV show that allows entrepreneurs to pitch their business ideas to a panel of wealthy investors, known as sharks. Shark Tank India is a spin-off of the original show, featuring Indian entrepreneurs and investors. I will try to explore the Shark Tank India dataset and conduct exploratory data analysis to gain insights into the showâ€™s dynamics.

#### Reading the CSV file (Shark Tank India dataset) using Pandas

```python
import pandas as pd
sharkTankIndia_df = pd.read_csv('Shark Tank India.csv')

# cleaning of the data 
cleaning = sharkTankIndia_df[sharkTankIndia_df['Yearly Revenue']==-1]
for ele in cleaning.index:
    sharkTankIndia_df.at[ele,'Yearly Revenue'] = 0

# To get the number of rows and columns in the dataset.
sharkTankIndia_df.shape
#  we can see that there are 53 columns and 320 rows in the dataset.
```

#### Making seperate data frames for both the seasons (Season 1 and Season 2)

```python
sharkTankIndiaS1_df = sharkTankIndia_df[sharkTankIndia_df['Season Number']==1]
sharkTankIndiaS2_df = sharkTankIndia_df[sharkTankIndia_df['Season Number']==2]
```

### Problem statement 1 -
##### What were the Industry wise number of startups in both the seasons?

```python
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
%matplotlib inline

# Different types of industries that came in the 1st season

a = 0  # variable 'a' stores the number of bins to be made in the histogram related to types of industries. 
for ele in sharkTankIndiaS1_df.groupby('Industry')['Industry'].count():
    a+=1

# Different types of industries that came in the 2nd season

b = 0  # variable 'b' stores the number of bins to be made in the histogram related to types of industries. 
for ele in sharkTankIndiaS2_df.groupby('Industry')['Industry'].count():
    b+=1

#Plotting multiple charts in a Grid.

fig, axes = plt.subplots(1,2, figsize = (12,4))
plt.tight_layout(pad=2)
axes[0].set_title('Industries in Season 1')
axes[1].set_title('Industries in Season 2')
g1 = sns.histplot(sharkTankIndiaS1_df['Industry'], bins=a, ax=axes[0]);
g2 = sns.histplot(sharkTankIndiaS2_df['Industry'], bins=b, ax=axes[1]);
g1.tick_params(labelrotation=90);
for i in g1.containers:
    g1.bar_label(i,)
g2.tick_params(labelrotation=90);
for i in g2.containers:
    g2.bar_label(i,)
```

##### Thus, it can be clearly seen that the **Food Industry** in both the seasons were a major participant.! This simply owns to the fact that **Food** is the spine of our Civilization, so whatever be the State, people will eat and Companies will emerge and facilitate newer and better ways to eat. Also, Food is a concept that is well tested and implemented so people are still getting funds in the same domain.

### Problem statement 2 -
##### What were the State wise number of startups in both the seasons?

```python
# Different States from which pitchers came in the 1st season

a = 0  # variable 'a' stores the number of bins to be made in the histogram related to different states in season 1. 
for ele in sharkTankIndiaS1_df.groupby('Pitchers State')['Pitchers State'].count():
    a+=1

# Different States from which pitchers came in the 2nd season

b = 0  # variable 'b' stores the number of bins to be made in the histogram related to different states in season 2. 
for ele in sharkTankIndiaS2_df.groupby('Pitchers State')['Pitchers State'].count():
    b+=1

#Plotting multiple charts in a Grid.

fig, axes = plt.subplots(1,2, figsize = (12,4))
plt.tight_layout(pad=2)
axes[0].set_title('State-wise Startups in Season 1')
axes[1].set_title('State-wise Startups in Season 2')
g1 = sns.histplot(sharkTankIndiaS1_df['Pitchers State'], bins=a, ax=axes[0]);
g2 = sns.histplot(sharkTankIndiaS2_df['Pitchers State'], bins=b, ax=axes[1]);
g1.tick_params(labelrotation=90);
for i in g1.containers:
    g1.bar_label(i,)
g2.tick_params(labelrotation=90);
for i in g2.containers:
    g2.bar_label(i,)
```

##### As far as State-wise Startups are concerned, **Maharashtra** leads the chart, followed by Delhi and Karnataka in both the seasons. This can be attributed to the fact that Maharashtra has a vibrant startup ecosystem, and also **Mumbai's** good connectivity with other countries, presence of prominent VC firms such as Sequoia Capital and Matrix Partners, rising concept of co-working space and a huge customer base, makes the city an attractive destination for startups.  The presence of **IIT Bombay**, which provides easy access to tech talent, has led to emergence of Maharashtra as a startup hub in Mumbai.

### Problem statement 4-
##### What was the average age of the pitchers in both the seasons?

```python
#  Average age of the pitchers in season 1 and season 2-
ageDataS1 = sharkTankIndiaS1_df.groupby('Pitchers Average Age')[['Pitchers Average Age']].count()
ageDataS2 = sharkTankIndiaS2_df.groupby('Pitchers Average Age')[['Pitchers Average Age']].count()
fig, axes = plt.subplots(1,2, figsize = (12,4))
plt.tight_layout(pad=4)
g1 = sns.barplot(data=ageDataS1, x=ageDataS1.index, y=ageDataS1['Pitchers Average Age'], ax=axes[0]);
g2 = sns.barplot(data=ageDataS2, x=ageDataS2.index, y=ageDataS2['Pitchers Average Age'], ax=axes[1]);
g1.set_title('Average age of Pitchers in Season 1');
g2.set_title('Average age of Pitchers in Season 2');
g1.set_ylabel('Count');
g2.set_ylabel('Count');
for i in g1.containers:
    g1.bar_label(i,)
for i in g2.containers:
    g2.bar_label(i,)
```

##### Thus, from the above graph we can clearly see that about **65%** of the pitchers in season 1 and about **80%** of the pitchers in season 2 are **Middle** aged. Now, its a perception that starting up at 30s, 40s or 50s is too late, but perceptions are not necessarily spot on, and definitely not in this case. This **dominance of Middle aged** presenters in **Shark Tank India** can be attributed to various factors, one of them being, when one is fresh out of college, one might have education and great ideas, but unless one is born into a business family, the necessary human and financial capital would take time to earn. As one goes through the years of working in the industry one likes, one learns things hands-on, makes connections and earns money. Eventually when one is in his/her early to mid-40s, one takes the entrepreneurial plunge, has the resources handy. Long-term cultivation of networks, and partnerships gives a person credibility and reputation, which makes launching a business simpler and smoother..! After gaining some experience in the interested field, one can definitely startup. One must understand that **Mark Zuckerberg** is an Exception and not a Benchmark, thus we can clearly see the number of youth pitchers are even less than half as compared to experienced middle aged pitchers..!

#### Investment Analysis of Ashneer

```python
import random
ashneerInvAmt = sharkTankIndia_df[sharkTankIndia_df["Ashneer Investment Amount"]>0]['Ashneer Investment Amount']
ashneerInvEq = sharkTankIndia_df[sharkTankIndia_df["Ashneer Investment Amount"]>0]['Ashneer Investment Equity']
ashneerInvComp = sharkTankIndia_df[sharkTankIndia_df["Ashneer Investment Amount"]>0][['Startup Name','Industry','Number of sharks in deal','Total Deal Debt']]
fig, axes = plt.subplots(2,2,figsize=(14,12))
plt.tight_layout(pad=9)
# Plotting Ashneer's Investments
ax1 = sns.barplot(x=ashneerInvAmt, y=ashneerInvComp['Startup Name'], width =0.6, ax=axes[0,0])
ax1.set_title("Investments done by Ashneer (in lakhs)")
for i in ax1.containers:
    ax1.bar_label(i,)
# Plotting Ashneer's equity
ax2 = sns.barplot(x=ashneerInvEq, y=ashneerInvComp['Startup Name'], width =0.6, ax=axes[0,1])
ax2.set_title("Ashneer investment equity (in%)")
for i in ax2.containers:
    ax2.bar_label(i,)
# Plotting Ashneer's Industry wise investment
ashneerIndustry = ashneerInvComp.groupby('Industry').count()
data = ashneerIndustry['Startup Name']
keys = ashneerIndustry.index
explode = []
for ele in range(0,len(ashneerIndustry.index)):
    explode.append(random.uniform(0.1,0.15))
ax3 = plt.pie(data, labels=keys, explode=explode, autopct='%.0f%%')
plt.title('Industries in which Ashneer Invested');
# plt.show()
ax4 = sns.barplot(x=ashneerInvComp['Number of sharks in deal'], y=ashneerInvComp['Startup Name'], width =0.6, ax=axes[1,0])
ax4.set_title("Number of sharks in deal")
for i in ax4.containers:
    ax4.bar_label(i,)
```

#### Observing the plots above, we can elucidate Ashneer's Investment
